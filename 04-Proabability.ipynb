What is the probability of getting a spade with the draw of a single card from a deck of 52?

```python
# sample space
total_number_of_cards = 52

# favorable outcomes
number_of_favorable_cards = 13

P_heads = number_of_favorable_cards/total_number_of_cards
print(P_heads)
```
